````python
def initialize_inventory(products):
    """
    This function takes as input a list of products and creates an inventory dictionary, using the products in the list as key, and for the value 
    it ask the user for the quantity of each product.
    It also creates a counter to sum all the products

    This function returns two values, first the inventory dictionary and second the total product counter
    """

    inventory = {}
    total_products = 0

    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                inventory[product] = int(input(f"Please enter the quantity of {product}:"))
                if inventory[product] < 0:
                    raise ValueError
                valid_quantity = True
            except:
                print("Quantity must be a positive number. Please enter a valid quantity.")
    total_products += inventory[product]
    
    return inventory, total_products

#products = ["t-shirt", "mug", "hat", "book", "keychain"]

#inventory , total_products = initialize_inventory(products)
````

````python
def calculate_total_price(customer_orders):
    
    product_prices = []
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Please enter the price for {product}"))
                if price < 0:
                    raise ValueError
                product_prices.append(price)
                valid_price = True
            except:
                print("Price must be a positive number. Please enter a valid price.")   
    total_price = sum(product_prices)
    return total_price

#customer_orders = ["mug", "hat"]
#calculate_total_price(customer_orders)
````

````python
def get_customer_orders(products, inventory):
    
    customer_orders = set ()
    valid_product = False
    while not valid_product:
        try:
            number_of_products = int(input("How many items do you want to buy: "))
            if number_of_products < 1:
                raise ValueError
            valid_product = True
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número positivo")
    
    while len(customer_orders) < number_of_products:
        try:
            product = input("Please enter your order: ")
            if product in inventory and inventory[product] > 0:
                customer_orders.add(product)
            if inventory[product] == 0:
                raise ValueError
        
        except ValueError:
            print("The item selected is out of stock, please select another item")
        
        except KeyError:
            print("Item not recognized, please select another item")
````

````python
def update_inventory(customer_orders, inventory):

    inventory = {item : inventory[item]-1 if item in customer_orders else inventory[item] for item in inventory}
    inventory = {item : stock for item, stock in inventory.items() if stock > 0}
    inventory["total"] -= len(customer_orders)
    return inventory

def calculate_order_statistics(customer_orders, inventory):

    #products_ordered = len(customer_orders)
    #percentage = (products_ordered / total_products) * 100
    return (len(customer_orders), (len(customer_orders)/inventory["total"])*100)

def print_order_statistics(order_statistics):

    print(f"Order Statistics:\nTotal Products Ordered: {order_statistics[0]}\nPercentage of Products Ordered: {order_statistics[1]}%")

def print_updated_inventory(inventory):

    [print(f'{item}: {stock}') for item, stock in inventory.items()]
````
